```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

#reading the csv file
df=pd.read_csv('HR.csv')
df
```

```python
#removing unwanted columns
data=df.drop('last_evaluation',axis=1)
data
```

```python
#checking for duplicate values
sum(data.duplicated(data.columns))
```

```python
#removing duplicates
data=data.drop_duplicates(data.columns)
data
```

```python
sum(data.duplicated(data.columns))
```

```python
data.info()
```

```python
#transforming data type
data['Department']=data['Department'].astype('category')
data['salary']=data['salary'].astype('category')
data.info()
```

```python
#checking for null values
data.isnull().sum()
```

```python
#no null values
data.describe()
```

```python
#using heatmap for checking correlation of attributes
import seaborn as sns
corr=data.corr()
sns.heatmap(corr)
```

```python
#satisfaction level range using boxplot
plt.figure(figsize=(9,2))
sns.boxplot(x='satisfaction_level',data=data)
plt.show()
```

```python
#using label encoder for Department and salary columns
from sklearn.preprocessing import LabelEncoder as LE
le=LE()
data['Department']=le.fit_transform(data['Department'])
data['salary']=le.fit_transform(data['salary'])
```

```python
#train_test split
from sklearn.model_selection import train_test_split as tts
ft=data[['average_montly_hours','satisfaction_level','salary']]
label=data.left
x_train, x_test,y_train, y_test=tts(ft,label)
```

```python
#model accuracy
from sklearn.linear_model import LogisticRegression as lr

model=lr()
model.fit(ft,label)
model.score(ft,label)
```

```python
#finally predicting the values
print(label[11995:11996])
model.predict(ft[11995:11996])
```
